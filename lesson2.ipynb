```python
# Grant olishga moslik
# Real hayotdagi vaziyat
# O‘quv markazi qaysi talabalar grant olishga loyiq ekanini avtomatik aniqlamoqchi.
# Vazifa
# Talabaning o‘rtacha bali va davomat foizi berilgan. U grant olishga loyiq yoki yo‘qligini
# aniqlang.
# Ball = 80
# Davomat = 75
# Kirish ma’lumotlari
# O‘rtacha ball (0–100)
# Davomat foizi (0–100)
# Chiqish
# "Mos" yoki "Mos emas"

ball = int(input("Ball: "))
davomat = int(input("Davomat: "))
if ball >= 80 and davomat >= 75:
  print("Mos")
else:
  print("Mos emas")
```

```python
# Onlayn do‘kon chegirmasi
# Real hayotdagi vaziyat
# Internet do‘kon xarid summasiga qarab chegirma beradi.
# Vazifa
# Xarid summasiga qarab yakuniy to‘lovni hisoblang.
# Masalan:
# 500 000 dan kam — chegirma yo'q
# 500 000 – 1 000 000 — 10% chegirma
# 1 000 000 dan yuqori — 20% chegirma
# Kirish
# Xarid summasi
# Chiqish
# To‘lanadigan yakuniy summa

summa = int(input("Xarid summasini kiriting: "))
if summa < 500000:
  print("Chegirma yo'q")
elif summa >= 500000 and summa < 1000000:
  print(f"10% chegirma, umumiy: {summa * 0.9}")
else:
  print(f"20% chegirma, umumiy: {summa * 0.8}")
```

```python
# Eng ko‘p sotilgan mahsulot
# Real vaziyat
# Do‘kon menejeri qaysi mahsulot eng ko‘p sotilganini bilmoqchi.
# Vazifa
# Berilgan mahsulotlar ro‘yxatidan eng ko‘p takrorlanganini toping.
# Kirish
# ["Olma"
# ,
# "Banan"
# ,
# "Olma"
# ,
# "Apelsin"
# ,
# "Olma"]
# Chiqish
# "Olma"

mevalar = ["Olma", "Banan", "Olma", "Apelsin", "Olma"]
print(f"Eng ko'p mahsulot: {max(mevalar, key = mevalar.count)}")
```

```python
# Takroriy foydalanuvchini aniqlash
# Real vaziyat
# Tizimda bir xil foydalanuvchi nomi qayta ro‘yxatdan o‘tmasligi kerak.
# Vazifa
# Berilgan foydalanuvchi nomlari ro‘yxatida takror bor yoki yo‘qligini aniqlang.
# Kirish
# ["ali"
# ,
# "Sardor"
# ,
# "ali"]
# Chiqish
# True

users = ["ali", "Sardor", "ali"]
print(len(users) != len(set(users)))
```

```python
# Sinf natijalarini tahlil qilish
# Real vaziyat
# O‘qituvchi sinf natijalarini tezda tahlil qilmoqchi.
# ballar = [72, 88, 95, 61, 79, 85]
# Vazifa
# Berilgan ballar bo‘yicha:
# Eng yuqori ball
# Eng past ball
# O‘rtacha ball
# ni hisoblang.

ballar = [72, 88, 95, 61, 79, 85]
eng_yuqori = max(ballar)
eng_past = min(ballar)
ortacha = sum(ballar)/len(ballar)
print(f"Eng yuqori: {eng_yuqori}")
print(f"Eng past: {eng_past}")
print(f"Eng yuqori: {ortacha}")
```

```python
# O‘quvchilar reytingi
# Real vaziyat
# Maktab o‘quvchilarni balliga qarab tartiblamoqchi.
# Vazifa
# Berilgan ballarni katta qiymatdan kichik qiymatga qarab tartiblang.

ballar = [72, 88, 95, 61, 79, 85]
print(sorted(ballar, reverse = True))
```

```python
# Talabalar bazasi
# Real vaziyat
# Maktab talabalar ma’lumotlarini saqlaydi.
# Vazifa
# Talaba ismi berilganda uning bahosini qaytaring.
# talabalar = { "Ali": 85,
# "Sardor": 92,
# "Malika": 78}

talabalar = { "Ali": 85,
              "Sardor": 92,
              "Malika": 78 }
name = input("Ism kiriting: ")
print(talabalar.get(name, "Bunday ism yo'q!"))
```

```python
# So‘zlar chastotasi
# Real vaziyat
# Qidiruv tizimlari matnda so‘zlar necha marta ishlatilganini hisoblaydi.
# Vazifa
# Berilgan matndagi har bir so‘z nechta marta takrorlanganini aniqlang

text = input("Matnni kiriting: ")
info = {}
for word in text.split( ):
  if word in info:
    info[word] += 1
  else:
    info[word] = 1
for i in info:
  print(f"{i} so'zi {info[i]} marta qatnashgan!")
```

```python
# Talaba baholash tizimi
# Real vaziyat
# Maktab avtomatik baholash tizimi yaratmoqchi.
# Vazifa
# Quyidagi funksiyalarni yarating:
# O‘rtacha hisoblash
# Harfli baho berish (A, B, C…)
# Hisobot chiqarish
# O’rtacha ball > 90 - A baho
# O'rtacha ball > 80 - B baho
# O'rtacha ball > 70 - C baho
# O’rtacha ball > 60 - D baho
# qolganlari - F
# Misol
# Talaba: Sardor O'rtacha ball: 86.2 Baho: B

name = input("Ism kiriting: ")
score = float(input("Baho kiriting: "))
if score > 90:
  print(f"Talaba: {name} O'rtacha ball: {score} Baho: A")
elif score > 80:
  print(f"Talaba: {name} O'rtacha ball: {score} Baho: B")
elif score > 70:
  print(f"Talaba: {name} O'rtacha ball: {score} Baho: C")
elif score > 60:
  print(f"Talaba: {name} O'rtacha ball: {score} Baho: D")
else:
  print(f"Talaba: {name} O'rtacha ball: {score} Baho: F")
```
